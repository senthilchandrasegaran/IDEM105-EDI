```python
def count_matches(text, pattern):
    start_star = False
    end_star = False
    if pattern.startswith('*') :
        pattern = r"[A-Za-z]*" + pattern[1:]
        start_star = True
        
    if pattern.endswith('*') :
        pattern = pattern[:-1] + r"[A-Za-z]*"
        end_star = True
        
    # the order of this next statement matters 
    # as it can be done only after the above two are executed
    if '*' in pattern :
        pattern.replace('*', ".*")
    
    if start_star and end_star :
        m = pattern
    elif start_star :
        m = pattern + r"\b"
    elif end_star :
        m = r"\b" + pattern
    else :
        m = r"\b" + pattern + r"\b"
    
    matches = re.findall(m, text)
    return matches
```
