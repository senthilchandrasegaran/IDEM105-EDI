```python
import nltk
import datetime
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
transcript_df = pd.read_csv("./data/The Miracle Sudoku_5sec_timestamps.csv", delimiter=";")
transcript_df.to_excel("./data/sudoku_transcript_timestamps.xlsx")
transcript_df.sample(5)
```

```python
transcript_df['Length'] = transcript_df['Utterance'].apply(len)
transcript_df.sample(5)
```

```python
plt.figure(figsize=(24, 4), dpi=600)
sns.set_style('ticks')
sns.set_context('paper', font_scale=0.5)
g = sns.barplot(data=transcript_df, x='Timestamp', y='Length', facecolor='gray')
# g.set_xlim([time_to_ftime('00:00:05'), transcript_df['Time'].to_list()[-1]])
g.xaxis.set_ticks(transcript_df['Timestamp'].to_list())
sns.despine(top=True, right=True, left=True, bottom=False)
h = g.set_xticklabels(g.get_xticklabels(), rotation=90,  ha='center', va='top')
plt.savefig('./plots/sudoku_5_sec.pdf', bbox_inches='tight')
```

```python
transcript_30_df = pd.read_csv("./data/miracle-sudoku-30sec.csv", delimiter=";")
transcript_30_df.to_excel("./data/miracle-sudoku-30sec.xlsx")
transcript_30_df.sample(5)
```

```python
transcript_30_df['Length'] = transcript_30_df['Utterance'].apply(len)
transcript_30_df.sample(5)
```
